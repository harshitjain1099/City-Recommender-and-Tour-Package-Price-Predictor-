# visit_place_summary

```python
copy_df['visit_place_summary'].value_counts()
```

```python
copy_df['visit_place_summary'].value_counts().shape
```

```python
import string
```

```python
def remove_punctuation(text):
    
    for i in string.punctuation:
        if i in text:
            text = text.replace(i,'')
    return text
```

```python
copy_df['visit_place_summary'] = copy_df['visit_place_summary'].apply(remove_punctuation)
```

```python
copy_df['visit_place_summary'] = copy_df['visit_place_summary'].str.lower()
```

```python
copy_df['visit_place_summary'].value_counts()
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize 
```

```python
copy_df['tokenized_summary'] = copy_df['visit_place_summary'].apply(word_tokenize)
```

```python
copy_df.head(10)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    
    L = []
    for word in text:
        if word not in stopwords.words('english'):
            L.append(word)
    return L
```

```python
copy_df['tokenized_summary'] = copy_df['tokenized_summary'].apply(remove_stopwords)
```

```python
copy_df.head(10)
```

```python
copy_df['visit_place_summary'] = copy_df['tokenized_summary'].apply(lambda x: " ".join(x))
```
