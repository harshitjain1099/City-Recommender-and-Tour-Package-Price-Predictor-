```python

final_df = pd.DataFrame()


for city in cities:
    
    url = 'https://www.holidify.com/places/{}/sightseeing-and-things-to-do.html'.format(city)
    
    
    response = requests.get(url)
    html_contents = response.text
    
    
    soup = BeautifulSoup(html_contents, 'lxml')
    
    
    packages = soup.find_all('div', class_="col-12 col-md-6 pr-md-3")
    
    
    visit_places = []
    rating = []
    places = []
    all_about = []
    
    
    for package in packages:
        visit_places.append(package.find('h3').text.strip() if package.find('h3') else None)

        rating_element = package.find('span', class_='rating-badge')
        rating.append(rating_element.text.strip() if rating_element else None)

        places.append(package.find('p', class_='objective').text.strip() if package.find('p', class_='objective') else None)

        all_about.append(package.find('p', class_='card-text').text.strip() if package.find('p', class_='card-text') else None)
    
    
    df = pd.DataFrame({
        'visit_places': visit_places,
        'rating': rating,
        'city_names':places,  
        'visit_place_summary': all_about
    })
    

    final_df = pd.concat([final_df, df], ignore_index=True)

```
