```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from requests.exceptions import ConnectionError, Timeout

final_df_3 = pd.DataFrame()

# Define the number of pages you want to scrape for each city
num_pages = 5  # Adjust this number based on how many pages each city has

for city in cities:
    for page in range(0, num_pages + 1):  # Loop through each page
        url = f'https://www.holidify.com/places/{city}/packages.html?pageNum={page}'
        
        try:
            response = requests.get(url, timeout=10)  # Set a timeout of 10 seconds
            response.raise_for_status()  # Check if the request was successful
        except ConnectionError:
            print(f"Connection aborted for {city} on page {page}. Skipping this page.")
            continue
        except Timeout:
            print(f"Request timeout for {city} on page {page}. Skipping this page.")
            continue
        except Exception as e:
            print(f"Error for {city} on page {page}: {e}")
            continue
        
        html_contents = response.text
        soup = BeautifulSoup(html_contents, 'lxml')
        
        # Find all packages on the current page
        new_packages = soup.find_all('div', class_="row no-gutters inventory-card")

        # Lists to store the data for each page
        package_names, trip_durations, city_lists, package_prices, personss = [], [], [], [], []
        
        for new_package in new_packages:
            package_names.append(new_package.find("h3", class_="text-truncate").text.strip() if new_package.find("h3", class_="text-truncate") else None)
            trip_durationss = new_package.find('p', class_='trip-duration')
            trip_durations.append(trip_durationss.text.strip() if trip_durationss else None)
            city_lists.append(new_package.find('p', class_='places-covered').text.strip() if new_package.find('p', class_='places-covered') else None)
            package_prices.append(new_package.find('p', class_='price').text.strip() if new_package.find('p', class_='price') else None)
            personss.append(new_package.find('p', class_='price-desc').text.strip() if new_package.find('p', class_='price-desc') else None)

        # Create a DataFrame for the current city and page
        df = pd.DataFrame({
            'package_name': package_names,
            'duration': trip_durations,
            'city_list': city_lists,
            'price': package_prices,
            'persons': personss
        })
        
        # Concatenate with the final DataFrame
        final_df_3 = pd.concat([final_df_3, df], ignore_index=True)

        # Sleep to avoid hitting the server too frequently
        time.sleep(3)

# Optionally, save the DataFrame to a CSV file
# final_df_3.to_csv('holidify_packages.csv', index=False)
```
