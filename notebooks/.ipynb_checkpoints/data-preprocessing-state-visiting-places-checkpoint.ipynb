# turist_attration

```python
copy_df['turist_attration'].value_counts()
```

```python
copy_df['turist_attration'].value_counts().shape
```

```python
 copy_df['turist_attration'] =  copy_df['turist_attration'].apply(lambda name : re.sub(r'\s+', ' ', str(name)).strip()).str.lower().str.split("in").str.get(1)
```

```python
 copy_df['state_name'] = copy_df['turist_attration'].str.split('tourist').str.get(0).apply(lambda name : re.sub(r'\s*\d+\s*', ' ', str(name)).strip())
```

```python
copy_df['no_of_turist_attration'] = copy_df['turist_attration'].str.split('tourist').str.get(0).apply(lambda name : re.sub(r'\b[a-zA-Z]+\b', ' ', str(name)).strip())
```

```python
copy_df.head(10)
```

```python
copy_df.drop(columns=['turist_attration'],inplace=True)
```
