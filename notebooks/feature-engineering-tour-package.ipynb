```python
def categories_by_price(price):
    if 0 <= price <= 15000:
        return "Budget"
    elif 15000 < price <= 30000:
        return "Standard"
    elif 30000 < price <= 50000:
        return "Premium"
    elif price > 50000:
        return "Luxury"
```

```python
def categories_by_package(tokenized_package_name):
    for word in tokenized_package_name:
        if word in adventure_packages:
            return 'adventure_package'
        elif word in pilgrimage_packages:
            return "Spiritual Packages"
        elif word in honeymoon_packages:
            return "honeymoon_package"
        elif word in nature_packages:
            return "nature_package"
        elif word in cultural_packages:
            return "cultural_package"
        elif word in luxury_packages:
            return "luxury_package"
        elif word in family_packages:
            return "family_package"
        elif word in budget_packages:
            return "budget_package"
        elif word in wellness_packages:
            return "wellness_package"
        elif word in offbeat_packages:
            return "offbeat_package"
        elif word in weekend_packages:
            return "weekend_package"
        elif word in beach_holidays:
            return "beach_holiday"
    
    
    return "holiday_package"
```

```python
def get_accommodation_facility(budget, days):
    if budget == 'budget':
        if days == 'short trip':
            return 'Low-Cost Apartment'
        elif days == 'medium trip':
            return 'Budget Hotel'
        elif days == 'long trip':
            return 'Hostel Room'
        elif days == 'extended trip':
            return 'Low-Cost shared Stay'
    elif budget == 'standard':
        if days == 'short trip':
            return 'Serviced Apartment'
        elif days == 'medium trip':
            return '3-Star Hotel'
        elif days == 'long trip':
            return 'Standard Hotel'
        elif days == 'extended trip':
            return 'Standard Hostel'
    elif budget == 'premium':
        if days == 'short trip':
            return 'Premium Serviced Apartment'
        elif days == 'medium trip':
            return '5-Star Hotel'
        elif days == 'long trip':
            return '4-Star Hotel'
        elif days == 'extended trip':
            return 'Premium Hotel'
    elif budget == 'luxury':
        if days == 'short trip':
            return 'Private Villa/Luxury Apartment'
        elif days == 'medium trip':
            return 'High-End Luxury Resort'
        elif days == 'long trip':
            return '5-Star Resort'
        elif days == 'extended trip':
            return 'Luxury Guesthouse/Resort'
    return 'Other Accommodation'

```

```python
# Apply the function row-wise and pass both 'categories_by_price' and 'categories_by_days' columns
df['accommodation_facility'] = df.apply(lambda row: get_accommodation_facility(row['categories_by_price'], row['categories_by_days']), axis=1)
```
