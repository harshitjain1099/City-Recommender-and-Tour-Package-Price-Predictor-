```python
visit_places = []
rating = []
turist_attration = []
all_about =[]
best_time = []

for i in package:
    visit_places.append(i.find('h3',class_="card-heading").text.strip() if i.find('h3',class_="card-heading") else None)

    rating_element = i.find('span', class_='rating-badge')
    rating.append(rating_element.text.strip() if rating_element else None)
    
    turist_attration.append(i.find('p',class_="objective").text.strip() if i.find('p',class_="objective") else None)

    all_about.append(i.find('p', class_='card-text').text.strip() if i.find('p', class_='card-text') else None)
    
    best_time.append(i.find('p', class_='mb-3').text.strip() if i.find('p', class_='mb-3') else None)

    
    
    df=pd.DataFrame({'visit_places':visit_places,
    'rating':rating,
    'turist_attration' :turist_attration,         
    'visit_place_summary':all_about,
    'best_time':best_time
})
```

```python
final_df = pd.DataFrame()


for state in indian_states:
    
    url = "https://www.holidify.com/state/{}/top-destinations-places-to-visit.html".format(state)
    
    response = requests.get(url)
    html_contents = response.text
    
    
    soup = BeautifulSoup(html_contents, 'lxml')
    
    
    packages = soup.find_all('div', class_="card content-card")
    

    
    
    
    
    visit_places = []
    rating = []
    turist_attration = []
    all_about =[]
    best_time = []

    for package in packages:
        visit_places.append(package.find('h3',class_="card-heading").text.strip() if package.find('h3',class_="card-heading") else None)

        rating_element = package.find('span', class_='rating-badge')
        rating.append(rating_element.text.strip() if rating_element else None)
    
        turist_attration.append(package.find('p',class_="objective").text.strip() if package.find('p',class_="objective") else None)

        all_about.append(package.find('p', class_='card-text').text.strip() if package.find('p', class_='card-text') else None)
    
        best_time.append(package.find('p', class_='mb-3').text.strip() if package.find('p', class_='mb-3') else None)

    
    
    df=pd.DataFrame({'visit_places':visit_places,
              'rating':rating,
              'turist_attration' :turist_attration,         
              'visit_place_summary':all_about,
              'best_time':best_time
    })
    

    final_df = pd.concat([final_df, df], ignore_index=True)
```
